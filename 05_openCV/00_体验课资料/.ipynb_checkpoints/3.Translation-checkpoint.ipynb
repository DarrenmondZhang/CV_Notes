#  图片的平移

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
    
def imread(image):
    image = cv2.imread(image)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
```

```python
image = imread('image.jpg')
M = np.float32([[1,0,250],[0,1,500]]) # 向右移动250个像素点，向下移动500个像素点
shifted1 = cv2.warpAffine(image, M, (image.shape[1],image.shape[0]))
show(shifted1)
```

```python
image = imread('image.jpg')
M = np.float32([[1,0,-250],[0,1,-500]]) # 向左移动250个像素点，向上移动500个像素点
shifted = cv2.warpAffine(image, M, (image.shape[1],image.shape[0]))
show(shifted)
```
