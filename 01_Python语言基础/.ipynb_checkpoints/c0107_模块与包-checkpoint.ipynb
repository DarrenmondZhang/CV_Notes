```python
# f1(n)f2(lst,x),f3(a,d,n)

def f1(n):
    y = 1
    for i in range(1,n+1):
        y = y * i
    return y
# f1(n)

def f2(lst,x):
    while x in lst:
        lst.remove(x)
    return lst
# f2(lst,x)

def f3(a,d,n):
    an = a
    s = 0
    for i in range(n-1):
        an = an + d
        s = s + an
    return s
# f3(a,d,n)
# testmodel2
```
