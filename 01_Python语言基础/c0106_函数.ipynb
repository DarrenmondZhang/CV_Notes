```python
def fn(st):
    n_alpha = 0
    n_space = 0
    n_digit = 0
    n_qt = 0
    for i in st:
        if i.isalpha():
            n_alpha += 1
        elif i.isdigit():
            n_digit += 1
        elif i.isspace():
            n_space += 1
        else:
            n_qt += 1
    print('%i %i %i' %(n_alpha, n_space, n_digit))
fn('sdd fs fd')
```
